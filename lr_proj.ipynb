```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Train, test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=111)
```

```python
X_train.shape, y_train.shape
```

```python
# standardization
from sklearn.preprocessing import StandardScaler

normalizer = StandardScaler()
```

```python
X_train = normalizer.fit_transform(X_train)
```

```python
import pickle
pickle.dump(normalizer, open('normalizer.pkl', 'wb'))
```

```python
X_test = normalizer.transform(X_test)
```

```python
 # Training the dataset 
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
# getting final coeff and intercept

lm.coef_, lm.intercept_
```

```python
# on which parameter is the model trained on
lm.get_params()
```

```python
X_pred = lm.predict(X_test)
```

```python
X_pred
```

```python
# plot scatter plot for the prediction
plt.scatter(x=X_pred, y=y_test)
plt.show()
```

```python
# prediction on the residual 

residuals = y_test - X_pred
```

```python
sns.displot(residuals, kind='kde')
plt.show()
```

```python
# Scatter plot with respect to prediction and reresiduals

plt.scatter(X_pred, residuals)
plt.show()
```

```python
# performance 
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(f'Mean squared error: {mean_squared_error(y_test, X_pred)}')
print(f'Mean absolute error: {mean_absolute_error(y_test, X_pred)}')
print(f'Root mean square error: {np.sqrt(mean_squared_error(y_test, X_pred))}')
```

# R square and adjusted R square

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, X_pred)
score
```
